## Soluzione

### Definizione della funzione `format_fasta()`

La funzione prende come argomento una stringa contenente un *header* `FASTA` e una sequenza (nucleotidica o proteica) e restituisce la sequenza in formato `FASTA`.

```python
def format_fasta(header, sequence):
    return header + '\n' + '\n'.join(re.findall('\w{1,80}', sequence))
```

**NOTA BENE**: supporre che l'*header* in input alla funzione abbia già il simbolo `>` all'inizio, ma non il simbolo `\n` alla fine.

### Parametri in input

```python
genetic_code_name = './genetic-code.txt'
input_file_name = './M10051.txt'
```

Importazione del modulo `re` per utilizzare le espressioni regolari.

### Lettura del file del codice genetico in una lista di stringhe `file_str_list`

```python
with open(genetic_code_name, 'r') as genetic_file:
    file_str_list = genetic_file.readlines()
```

### Costruzione del dizionario del codice genetico

Estrarre da `file_str_list` il dizionario `genetic_code_dict`:

- *chiave*: sequenza del codone
- *valore*: simbolo del corrispondente amminoacido

Costruire prima la lista `key_tuple_list` delle tuple *(chiave, valore)* di `genetic_code_dict` e in seguito costruire il dizionario con la funzione `dict()` a cui viene passata `key_tuple_list` come argomento.

```python
key_tuple_list = [(codon, row.rstrip().split(',')[0]) for row in file_str_list for codon in row.rstrip().split(',')[1:]]
```

```python
key_tuple_list
```

Costruire il dizionario con la funzione `dict()`.

```python
genetic_code_dict = dict(key_tuple_list)
```

```python
genetic_code_dict
```

### Lettura del file `EMBL` nell'unica stringa `file_str`

```python
with open(input_file_name, 'r') as input_file:
    file_str = input_file.read()
```

```python
print(file_str)
```

### Estrazione dell'identificatore univoco e dell'organismo relativo all'entry.

Estrarre dal *record* `ID`:

    ID   M10051; SV 1; linear; mRNA; STD; HUM; 4723 BP.
    
l'identificatore univoco e l'organismo e assegnarli alle variabili `identifier` e `organism`.
    
Estrazione dell'organismo.

```python
s = re.search('([\w\s]+;){5}\s+(\w+);', file_str, re.M)
organism = s.group(2)
```

```python
organism
```

Estrazione dell'identificatore.

```python
s = re.search('^ID\s+(\w+);', file_str, re.M)
identifier = s.group(1)
```

```python
identifier
```

### Estrazione della sequenza nucleotidica

Estrarre nella lista `seq_row_list` i *record*: 

    tgggggccgc ctcggagcat gacccccgcg ggccagcgcc gcgcgcctga tccgaggaga       120
    
della sequenza nucleotidica escludendo l'intero finale e mantenendo gli spazi iniziali e gli spazi prima dell'intero.

```python
seq_row_list = re.findall('^\W{2}(\D+)\d+', file_str, re.M)
```

```python
seq_row_list
```

Estrarre da `seq_row_list` la lista `seq_chunk_list` contenente i pezzi di lunghezza 10 della sequenza nucleotidica in modo tale che l'elemento `seq_chunk_list[i]` sia una lista annidata contenente i sei pezzi di sequenza relativi all'i-esimo *record* di `seq_row_list`.

**NOTA BENE**: fare attenzione al fatto che l'ultimo pezzo potrebbe essere più corto di 10 basi.

```python
seq_chunk_list = [re.findall('\w+', row) for row in seq_row_list]
```

```python
seq_chunk_list
```

Concatenare i pezzi della lista `seq_chunk_list` in un'unica stringa e assegnarla alla variabile `nucleotide_sequence`.

```python
nucleotide_sequence = ''.join([''.join(list_six_chunks) for list_six_chunks in seq_chunk_list])
```

```python
nucleotide_sequence
```

### Estrazione della sequenza della proteina

Estrarre il prefisso della proteina contenuto nel *record*:

    FT                   /translation="MGTGGRRGAAAAPLLVAVAALLLGAAGHLYPGEVCPGMDIRNNLT
    
e assegnarlo alla variabile `protein_prefix`.

```python
s = re.search('^FT\s+\/translation=\"(\w+)$', file_str, re.M)
protein_prefix = s.group(1)
```

```python
protein_prefix
```

Riconoscere gli altri *record*:

    FT                   RLHELENCSVIEGHLQILLMFKTRPEDFRDLSFPKLIMITDYLLLFRVYGLESLKDLFP
    
che contengono la sequenza della proteina e aggiungere i rimanenti pezzi alla lista `protein_row_list`.
    
**NOTA BENE**: attenzione all'ultimo:

    FT                   DGGSSLGFKRSYEEHIPYTHMNGGKKNGRILTLPRSNPS"

che contiene alla fine dei doppi apici `"`.

```python
protein_row_list = re.findall('^FT\s+([ACDEFGHIKLMNPQRSTVWY]+)\"?$', file_str, re.M)
```

```python
protein_row_list
```

Aggiungere in testa alla lista `protein_row_list` il prefisso trovato prima e concatenare tutti i pezzi della lista `protein_row_list` per ottenere la sequenza della proteina in un'unica stringa. Assegnare la sequenza della proteina alla variabile `protein_sequence`.

```python
protein_row_list[:0] = protein_prefix
protein_sequence = ''.join([''.join(chunk) for chunk in protein_row_list])
```

```python
protein_sequence
```

### Output della coding sequence (CDS) in formato `FASTA`

Estrarre dal *record*

    FT   CDS             139..4287
    
lo start e l'end della CDS sulla sequenza nucleotidica dell'mRNA.

```python
s = re.search('^FT\s+CDS\s+(\d+)..(\d+)$', file_str, re.M)
cds_start = s.group(1) 
cds_end = s.group(2)
```

```python
cds_start
```

```python
cds_end
```

Ottenere la coding sequence (CDS) come sottostringa della sequenza dell'mRNA.

```python
cds_sequence = nucleotide_sequence[int(cds_start)-1:int(cds_end)]
```

```python
cds_sequence
```

Produrre la sequenza della CDS in formato `FASTA` con il seguente *header*:

    >M10051-HUM; len = 4149; start = yes; end = yes
    
e assegnarla alla variabile `cds_sequence_fasta`.

```python
header = '>' + identifier + '-' + organism + '; len = ' + str(len(cds_sequence)) + ';'

if cds_sequence[:3] == 'atg':
    header = header + ' start = yes;'
else:
    header = header + ' start = no;'

if cds_sequence[-3:] in ['taa', 'tag', 'tga']:
    header = header + ' end = yes;'
else:
    header = header + ' end = no;'

cds_sequence_fasta = format_fasta(header, cds_sequence)
```

```python
print(cds_sequence_fasta)
```

### Output della distribuzione della frequenza dei codoni della CDS

Estrarre la lista dei codoni della CDS (stop codon escluso).

```python
codon_list = re.findall('\w{3}', cds_sequence)[:-1]
```

```python
codon_list
```

Costruire la lista `codon_frequency` di tuple *(codon, frequency)* elencate per frequenza decrescente.

```python
from collections import Counter

codon_counter = Counter(codon_list)
codon_frequency = codon_counter.most_common()
```

```python
codon_frequency
```

### Output della distribuzione della frequenza degli amminoacidi della proteina letta dal file `EMBL`

Estrarre la lista degli amminoacidi della proteina letta dal file `EMBL`.

```python
ammino_list = re.findall('\w', protein_sequence)
#In alternativa
ammino_list = list(protein_sequence)
```

```python
ammino_list
```

Costruire la lista `ammino_frequency` di tuple *(amminoacid, frequency)* elencate per frequenza decrescente.

```python
ammino_counter = Counter(ammino_list)
ammino_frequency = ammino_counter.most_common()
```

```python
ammino_frequency
```

### Validazione della sequenza della proteina letta dal file `EMBL`

Tradurre in proteina (tramite il codice genetico) la sequenza della CDS.

**Alternativa1:** traduzione dei codoni in `codon_list` e unione in unica stringa con il metodo `join()`.

```python
genetic_list = [genetic_code_dict[codon] for codon in codon_list]
cds_translation1 = ''.join(genetic_list)
```

```python
cds_translation1
```

**Alternativa2:** traduzione con la funzione `sub()` del modulo `re`.

Alla funzione:

    re.sub(my_expr, r_arg, my_string)

può essere passato come secondo argomento `r_arg` una funzione al posto della stringa di sostituzione. In questo caso `sub()` invochera' tale funzione passando come argomento l'oggetto restituito da ognuna delle operazioni di *searching* della espressione regolare `my_expr` sulla stringa `my_string`.

Il secondo argomento di `sub()` dovrà essere una funzione anonima definita secondo la seguente sintassi:

    lambda x: expression
    
dove `x` è l'argomento della funzione ed `expression` è l'espressione da valutare e il cui risultato verrà restituito.

**NOTA BENE**: ricordarsi di eliminare l'ultimo codone di stop.

```python
cds_translation2 = re.sub('\w{3}', lambda x: genetic_code_dict[x.group()], cds_sequence[:-3])
```

```python
cds_translation2
```

Verificare che le traduzioni ottenute nei due diversi modi siano uguali.

```python
cds_translation1 == cds_translation2
```

Verificare infine che la proteina letta dal file `EMBL` sia uguale a quella ottenuta traducendo la CDS.

```python
cds_translation1 == protein_sequence
```
