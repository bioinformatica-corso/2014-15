## Definizione della funzione che estrae i generi oppure i paesi di produzione relativi a un film

La funzione deve prendere in input un campo `genres` oppure `production_countries` e restituire la lista dei generi o dei paesi di produzione presenti nel campo.

Ad esempio dal campo:

    [{'id': 16, 'name': 'Animation'}, {'id': 35, 'name': 'Comedy'}]
    
deve ottenere la lista `['Animation', 'Comedy']`

mentre dal campo:

    [{'iso_3166_1': 'DE', 'name': 'Germany'}, {'iso_3166_1': 'US', 'name': 'United States of America'}]
    
deve ottenere la lista `['Germany', 'United States of America']`


```python
import ast
```

```python
def get_items(field_value):
    return [d['name'] for d in ast.literal_eval(field_value)]
```

## Estrazione dei 10 paesi che hanno prodotto più film

a) Costruire la lista dei paesi che hanno prodotto film, in modo che ciascun paese si presente nella lista un numero di volte pari al numero di film che ha prodotto.

```python
country_list = []

for (index, record) in df.iterrows():
    country_list.extend(get_items(record['production_countries']))
```

```python
country_list
```

b) Costruire il contatore

```python
from collections import Counter

country_rank_list = Counter(country_list).most_common()[:10]
```

```python
country_rank_list
```

```python
for (country, counter) in country_rank_list:
    print(country, 'produced', counter, 'movies')
```

## Trovare, per ogni genere, il film più popolare

a) Costruire il dizionario delle popolarità:

- **chiave**: genere
- **valore**: lista dei film associati al genere

Ogni film deve essere rappresentato dalla tupla `(popularity, original_title)`

**NB**: inserire il film solo se il valore del campo `popularity` è diverso da `NaN` (Not a Number) da controllare tramite la funzione `isnan()` di `numpy`.

```python
import numpy as np
```

```python
pop_dict = {}

for (index, record) in df.iterrows():
    if np.isnan(record['popularity']) == False:
        for gen in get_items(record['genres']):
            value = pop_dict.get(gen, [])
            value.append((record['popularity'], record['original_title']))
            pop_dict[gen] = value
```

```python
pop_dict
```

b) Produrre in output la lista dei film più popolari per genere.

```python
for key in pop_dict:
    movie_list = pop_dict[key]
    movie_list.sort(reverse = True)
    first_movie_title = movie_list[0][1]
    print('"'+first_movie_title+'"', 'for', key, 'movies')
```

## Trovare, per ogni paese, il genere preferito

a) Costruire il dizionario dei generi prodotti:

- **chiave**: paese di produzione P
- **valore**: dizionario (inner_dict) dei generi associati ai film prodotti dal paese P

Struttura dei dizionario inner_dict:
- **chiave**: genere G
- **valore**: numero di film prodotti dal paese P associati al genere G

```python
gen_dict = {}

#Per ogni record(film) nel data frame
for (index, record) in df.iterrows():
    
    #Recupero la lista dei paesi associati al film
    country_list = get_items(record['production_countries'])
    
    #Recupero la lista dei generi associati al film
    genre_list = get_items(record['genres'])
    
    if country_list != [] and genre_list != []:
        
        #Per ogni paese P che ha prodotto il film
        for country in country_list:
            #Recupero da gen_dict il valore (dizionario) associato alla chiave P --> inner_dict
            inner_dict = gen_dict.get(country, {})
            #Se viene recuperato un dizionario non vuoto, significa che la chiave P esiste già in
            #gen_dict, altrimenti significa che non esiste ancora e verrà quindi aggiunta a gen_dict
            
            #In entrambi i casi, il dizionario recuperato (vuoto oppure no) viene aggiornato con
            #i generi della lista genre_list (come chiavi). I conteggi, associati ai generi del
            #film corrente, vengono incrementati di 1.
            
            #In dettaglio: per ogni genere G della lista genre_list, recupero da inner_dict il valore (intero)
            #relativo alla chiave G, lo incremento di 1 e il risultato diventa il nuovo
            #valore relativo alla chiave G. Se G non è ancora presente in inner_dict, il metodo get
            #restituisce 0 (il conteggio parte da 0 per quel genere), e viene poi aggiornato a 1.
            for gen in genre_list:
                value = inner_dict.get(gen, 0) + 1
                inner_dict[gen] = value
     
            #Il valore relativo alla chiave P (paese) in gen_dict, deve essere aggiornato con il dizionario (aggiornato)
            #inner_dict
            gen_dict[country] = inner_dict
```

```python
gen_dict
```

b) Produrre in output il genere preferito di ogni paese di produzione.

```python
for country in gen_dict:
    pref_gen = max((counter, genre) for (genre, counter) in gen_dict[country].items())[-1]
    print(pref_gen, 'for', country)
```
